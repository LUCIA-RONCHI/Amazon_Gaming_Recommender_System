### Drop columns with a lot of NAs

```python
joined_data.drop(["date"],axis = 1, inplace=True)
joined_data.drop(["tech1"],axis = 1, inplace=True)
joined_data.drop(["details"],axis = 1, inplace=True)
joined_data.drop(["similar_item"],axis = 1, inplace=True)
joined_data.drop(["tech2"],axis = 1, inplace=True)
joined_data.drop(["reviewerName"],axis = 1, inplace=True)
joined_data.drop(["unixReviewTime"],axis = 1, inplace=True)
joined_data.drop(["image_x"],axis = 1, inplace=True)
joined_data.drop(["image_y"],axis = 1, inplace=True)
joined_data.drop(["price"],axis = 1, inplace=True)
```

### Convert Review to datetime

```python
#Change the review Time to datetime object
for i in range(len(joined_data["reviewTime"])):
    review_time=joined_data["reviewTime"][i]
    review_time=review_time.replace(",","")
    datetime_obj=datetime.strptime(review_time, "%m %d %Y").date()
    joined_data["reviewTime"][i]=datetime_obj
```
